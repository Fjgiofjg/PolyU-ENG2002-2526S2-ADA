```python
from task1 import fixNum

def get_fixNum(label=""):
    """Helper function to prompt the user and return a fixNum object."""
    print(f"--- Entering {label} fixed-point number ---")
    
    while True:
        try:
            a_input = input("Enter integer part a (can be negative): ")
            b_input = input("Enter fractional part b (must be positive): ")
            a = int(a_input)
            b = int(b_input)
            
            if b < 0:
                print("Invalid input! Fractional part 'b' cannot be negative. Please try again.")
                continue 
                
            return fixNum(a, b)
            
        except ValueError:
            print("Invalid input! You must enter valid integers. Please try again.")

def main():
    while True:
        print("\n=== Fixed-Point Number Calculator ===")
        print("1. Add two fixed-point numbers")
        print("2. Power of an integer (n)")
        print("3. Power of another fixed-point number")
        print("4. Exit")
        
        choice = input("Enter your choice (1-4): ")

        if choice == "4":
            print("Goodbye!")
            break

        # Task 1: Addition
        elif choice == "1":
            num1 = get_fixNum("first")
            num2 = get_fixNum("second")
            fixNum.calculate_basic_ops(choice, num1, num2)

        # Distinction Option 1: Power of an integer
        elif choice == "2":
            base = get_fixNum("base ")
            while True:
                try:
                    exponent = int(input("Enter an integer exponent n (can be negative): "))
                    result = base.power(exponent)
                    if result is not None:
                        print(f"Result: ({base}) ^ {exponent} = {result}")
                    break
                except ValueError:
                    print("Error: Exponent must be a valid integer. Please try again.")

        # Distinction Option 2: Power of another fixed-point number
        elif choice == "3":
            base = get_fixNum("base")
            exponent = get_fixNum("exponent")
            result = base.power(exponent)
            if result is not None:
                print(f"Result: ({base}) ^ ({exponent}) = {result}")
                
        else:
            print("Invalid choice. Please try again.")
            continue

if __name__ == "__main__":
    main()
```
