```python
class fixNum:
    def __init__(self, a=0, b=0):
        self.a = int(a)
        self.b = int(b)
        self.normalize()

    def normalize(self):
        if self.b >= 100:
            self.a += self.b // 100
            self.b = self.b % 100
        elif self.b < 0:
            borrow = (-self.b + 99) // 100
            self.a -= borrow
            self.b += borrow * 100

    def add(self, other):
        total = (self.a * 100 + self.b) + (other.a * 100 + other.b)
        return fixNum(total // 100, total % 100)

    def show(self):
        print(f"{self.a}.{self.b:02d}")


while True:
    print("\nFixed-Point Number Calculator")
    print("1. Add two fixed-point numbers")
    print("2. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        a1 = int(input("Enter first number integer part a: "))
        b1 = int(input("Enter first number fractional part b: "))
        a2 = int(input("Enter second number integer part a: "))
        b2 = int(input("Enter second number fractional part b: "))

        num1 = fixNum(a1, b1)
        num2 = fixNum(a2, b2)
        result = num1.add(num2)

        print("Result:")
        result.show()

    elif choice == "2":
        print("Goodbye.")
        break
    else:
        print("Invalid choice.")
```
