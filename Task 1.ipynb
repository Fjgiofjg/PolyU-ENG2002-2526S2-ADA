```python
from task1 import fixNum


def get_fixNum(label=""):
    a = int(input(f"Enter {label}integer part a: "))
    b = int(input(f"Enter {label}fractional part b: "))
    return fixNum(a, b)
while True:
    print("fixed-Point Number Calculator")
    print("1. Add two fixed-point numbers")
    print("2. Subtract two fixed-point numbers")
    print("3. Multiply two fixed-point numbers")
    print("4. Divide two fixed-point numbers")
    print("5. Power a fixed-point number")
    print("6. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        num1 = get_fixNum("first number ")
        num2 = get_fixNum("second number ")
        result = num1.add(num2)
        print("Result:", result)

    elif choice == "2":
        num1 = get_fixNum("first number ")
        num2 = get_fixNum("second number ")
        result = num1.sub(num2)
        print("Result:", result)

    elif choice == "3":
        num1 = get_fixNum("first number ")
        num2 = get_fixNum("second number ")
        result = num1.mul(num2)
        print("Result:", result)

    elif choice == "4":
        num1 = get_fixNum("first number ")
        num2 = get_fixNum("second number ")
        result = num1.div(num2)
        if result is None:
            print("Error: Division by zero.")
        else:
            print("Result:", result)

    elif choice == "5":
        num = get_fixNum("base number ")
        exponent = int(input("Enter integer exponent n: "))
        result = num.power(exponent)
        print("Result:", result)

    elif choice == "6":
        print("Goodbye.")
        break

    else:
        print("Invalid choice.")
```
