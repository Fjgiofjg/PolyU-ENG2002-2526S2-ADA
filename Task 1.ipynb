```python
# main.py
from task1 import fixNum

def get_fixNum(label=""):
    """Helper function to prompt the user and return a fixNum object."""
    print(f"--- Entering {label}fixed-point number ---")
    a = int(input("Enter integer part a (can be negative): "))
    b = int(input("Enter fractional part b (must be positive): "))
    return fixNum(a, b)

def main():
    while True:
        print("\n=== Fixed-Point Number Calculator ===")
        print("1. Add two fixed-point numbers")
        print("2. Subtract two fixed-point numbers")
        print("3. Multiply two fixed-point numbers")
        print("4. Divide two fixed-point numbers")
        print("5. Power of an integer (n)")
        print("6. Power of another fixed-point number")
        print("7. Exit")
        
        choice = input("Enter your choice (1-7): ")

        if choice == "7":
            print("Goodbye!")
            break
            
        if choice not in ["1", "2", "3", "4", "5", "6"]:
            print("Invalid choice. Please try again.")
            continue

        # Options 1-4 require two fixed-point numbers
        if choice in ["1", "2", "3", "4"]:
            num1 = get_fixNum("first ")
            num2 = get_fixNum("second ")
            
            if choice == "1":
                result = num1.add(num2)
                print(f"Result: {num1} + {num2} = {result}")
                
            elif choice == "2":
                result = num1.sub(num2)
                print(f"Result: {num1} - {num2} = {result}")
                
            elif choice == "3":
                result = num1.mul(num2)
                print(f"Result: {num1} * {num2} = {result}")
                
            elif choice == "4":
                result = num1.div(num2)
                if result is not None:
                    print(f"Result: {num1} / {num2} = {result}")

        # Option 5 requires a fixed-point base and an integer exponent
        elif choice == "5":
            base = get_fixNum("base ")
            try:
                exponent = int(input("Enter an integer exponent n (can be negative): "))
                result = base.power(exponent)
                if result is not None:
                    print(f"Result: ({base}) ^ {exponent} = {result}")
            except ValueError:
                print("Error: Exponent must be a valid integer.")

        # Option 6 requires two fixed-point numbers (base and exponent)
        elif choice == "6":
            base = get_fixNum("base ")
            exponent = get_fixNum("exponent ")
            result = base.power(exponent)
            if result is not None:
                print(f"Result: ({base}) ^ ({exponent}) = {result}")

if __name__ == "__main__":
    main()
```
